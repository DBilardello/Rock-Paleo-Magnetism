## Skewed Gaussian PDF Model Definition
$$
f(x) = \frac{A}{\sqrt{2\pi}\sigma} \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right) 
\left[ 1 + \operatorname{erf} \left( -\frac{\alpha(x - \mu)}{\sqrt{2} \sigma} \right) \right]
$$
- $A$: Amplitude  
- $\mu$: Center  
- $\sigma$: Dispersion (standard deviation)  
- $\alpha$: Skewness  
- $\operatorname{erf}$: Error function

````python
def skewed_gaussian(x, amplitude, center, sigma, skew):
    norm = amplitude / (np.sqrt(2 * np.pi) * sigma) * np.exp(-(x - center) ** 2 / (2 * sigma ** 2))
    erf_term = (1 + erf(-skew * (x - center) / (np.sqrt(2) * sigma)))
    return 1 * norm * erf_term

def model(x, params, n_components):
    y = np.zeros_like(x)
    for i in range(n_components):
        amp, cen, sig, skw = params[i*4:i*4+4]
        y += skewed_gaussian(x, amp, cen, sig, skw)
    return y
````

## Define the Interactive Widget with Field in mT

````python
def interactive_guess_matplotlib(n_components):
    """
    Interactive slider-based parameter guessing for skewed Gaussian fitting.

    Parameters:
        n_components (int): Number of skewed Gaussian components to fit.
                            This must be provided when calling the function in the following code cell,
                            example: interactive_guess_matplotlib(n_components=2)

    Displays center in log10(mT); internally converts to log10(ÂµT).
    """
    global initial_params, n_active_components
    n_active_components = n_components

    amp_scale = np.max(pdf)
    # Tableau 10 colors, skipping tab:blue (index 0)
    component_colors = [
    'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown',
    'tab:pink', 'tab:gray', 'tab:olive', 'tab:cyan'
    ]
    
    # === Generate initial parameters ===
    if 'initial_params' in globals() and len(initial_params) == n_components * 4:
        # Reuse prior values (in log10(ÂµT)) and convert back to log10(mT) for sliders
        defaults = []
        for i in range(n_components):
            amp = initial_params[i*4]
            center_log_uT = initial_params[i*4 + 1]
            sigma = initial_params[i*4 + 2]
            skew = initial_params[i*4 + 3]
            center_log_mT = center_log_uT - 3  # ÂµT â mT for display
            defaults.append([amp, center_log_mT, sigma, skew])
    else:
        # Generate new default guesses
        center_start_mT = 1.0   # log10(1 mT) = 1.0
        center_step = 0.2       # in log10(mT)
        defaults = []
        for i in range(n_components):
            center_log_mT = center_start_mT + i * center_step
            sigma = 0.5
            skew = 0.0
            amp = 0.00
            defaults.append([amp, center_log_mT, sigma, skew])
    # Internally convert center from log10(mT) â log10(ÂµT)
    params = []
    for amp, center_log_mT, sigma, skew in defaults:
        center_log_uT = center_log_mT + 3
        params.extend([amp, center_log_uT, sigma, skew])

    # === Set up layout using GridSpec ===
    slider_rows_per_component = 4
    slider_rows = n_components * slider_rows_per_component
    plot_rows = max(6, slider_rows // 2)
    buffer_rows = 2
    total_rows = plot_rows + buffer_rows + slider_rows

    fig = plt.figure(figsize=(9, 0.4 * total_rows + 3))
    height_ratios = [2] * plot_rows + [1] * buffer_rows + [1] * slider_rows
    gs = gridspec.GridSpec(total_rows, 1, figure=fig, height_ratios=height_ratios)

    # === Plot ===
    x_log_mT = x_log_uT - 3  # Shift to log10(mT)

    ax = fig.add_subplot(gs[:plot_rows])
    ax.scatter(x_log_mT, pdf, color='gray', label='Data (gradient)', alpha=0.75)
    model_line, = ax.plot(x_log_mT, model(x_log_uT, params, n_components), color='tab:blue', label='Mean Fit')
    comp_lines = []
    for i in range(n_components):
        color = component_colors[i % len(component_colors)]
        line, = ax.plot(x_log_mT, skewed_gaussian(x_log_uT, *params[i*4:i*4+4]), color=color)
        comp_lines.append(line)
    ax.set_xlabel("log B (mT)", fontsize=12)
    ax.set_ylabel("PDF (dM / dlogB)", fontsize=12)
    ax.legend(fontsize=10)

    # === Slider ranges ===
    param_names = ['Amp', 'Center', 'Sigma', 'Skew']
    ranges = [
        (0.0, 5 * amp_scale),   # Amplitude
        (0.0, 4),             # Center in logââ(mT)
        (0.05, 1.0),            # Sigma
        (-5.0, 5.0)             # Skew
    ]

    sliders = []
    for i in range(n_components):
        component_color = component_colors[i % len(component_colors)]
        for j in range(slider_rows_per_component):
            slider_idx = i * slider_rows_per_component + j
            ax_slider = fig.add_subplot(gs[plot_rows + buffer_rows + slider_idx])
            val_init = defaults[i][j]
            min_val, max_val = ranges[j]
            s = Slider(
                ax_slider,
                f'{param_names[j]} {i+1}: {min_val:.2f}',
                min_val,
                max_val,
                valinit=val_init,
                valfmt="%.2f",
                color=component_color, alpha=0.9
            )
            sliders.append(s)

    # === Update logic ===
    def update(val):
        slider_vals = [s.val for s in sliders]
        new_params = []
        for i in range(n_components):
            amp = slider_vals[i*4]
            center_log_mT = slider_vals[i*4 + 1]
            sigma = slider_vals[i*4 + 2]
            skew = slider_vals[i*4 + 3]
            center_log_uT = center_log_mT + 3  # Conversion
            new_params.extend([amp, center_log_uT, sigma, skew])
        model_line.set_ydata(model(x_log_uT, new_params, n_components))
        for i in range(n_components):
            comp_lines[i].set_ydata(skewed_gaussian(x_log_uT, *new_params[i*4:i*4+4]))
        fig.canvas.draw_idle()
        initial_params[:] = new_params

    for s in sliders:
        s.on_changed(update)

    initial_params = params.copy()
    update(None)
    plt.show()
````

## Setting bounds around parameters to make the optimization work best

````python
amp_scale = np.max(pdf)  # Redefine amp_scale here

# Example: define bounds per component
bounds = []
for i in range(n_active_components):
    amp_guess   = initial_params[i * 4]
    center_guess = initial_params[i * 4 + 1]
    sigma_guess = initial_params[i * 4 + 2]
    skew_guess  = initial_params[i * 4 + 3]

    # Amplitude: 0 to 1.25Ã initial guess
    bounds.append((0, max(amp_guess * 1.25, amp_scale * 0.05))) #bounds.append((0, amp_guess * 1.25))

    # Center: Â±0.3 around guess, within limits
    bounds.append((max(center_guess - 0.3, 2.5), min(center_guess + 0.3, 7.0))) #bounds.append((max(center_guess - 0.3, 3.0), min(center_guess + 0.3, 7.0)))

    # Sigma: 0.5Ã initial guess but capped at 0.05 to 2Ã initial guess
    #bounds.append((max(0.05, sigma_guess * 0.5), sigma_guess * 2))
    sigma_lower = max(0.05, sigma_guess * 0.5) # Ensures minimum is 0.05 even if set to zero
    sigma_upper = max(sigma_lower + 0.01, sigma_guess * 2)  # Ensure valid non-zero range
    bounds.append((sigma_lower, sigma_upper))

    # Skew: Â±2 from guess, capped at Â±5
    bounds.append((max(-5.0, skew_guess - 2.0), min(5.0, skew_guess + 2.0)))
````

## PDF Optimization
### Choosing Objective Function

To select which cost function to use for optimization:

## Set Number of Monte Carlo Simulations
- 500â1000 simulations are recommended for stability of the confidence bands.
- Fewer simulations (e.g. 100) can be used during development or for quicker runs, but may result in noisy uncertainty envelopes.

````python
monte_carlo_simulations= 500
````

````python
# === Objective function to minimize residuals ===
# Mean Squared Error (MSE) â interpretable on data scale
"""
def objective(params, x, pdf, n_active_components):
    y_fit = model(x, params, n_active_components)
    residuals = pdf - y_fit
    residual_cost = np.sum(residuals ** 2)
    return residual_cost
"""

# Relative L2 norm â unitless error relative to signal
def objective(params, x, pdf, n_active_components):
    y_fit = model(x, params, n_active_components)
    residuals = pdf - y_fit
    residual_cost = np.linalg.norm(residuals) / np.linalg.norm(pdf)
    return residual_cost


# === Calculate the intersection of two components using Brent's method ===
def find_right_intersection(x, comp1, comp2, mode1, mode2, buffer=0.1):
    mask = (x > mode1) & (x < (mode2 + buffer))
    x_masked = x[mask]
    
    if len(x_masked) < 2:
        print("Not enough points between peaks")
        return None
    
    try:
        intersection = brentq(lambda z: np.interp(z, x_masked, comp1[mask]) - np.interp(z, x_masked, comp2[mask]),
                              x_masked[0], x_masked[-1])
        return intersection
    except ValueError as e:
        print(f"Brentq failed: {e}")
        return None


# === Perform the constrained optimization with residual minimization ===
result = minimize(objective, initial_params, args=(x_log_uT, pdf, n_active_components), method='L-BFGS-B', bounds=bounds)

# Extract optimized parameters
optimized_params = result.x

# Calculate the best fit
best_fit = model(x_log_uT, optimized_params, n_active_components)


# === Function to find peak mode from fitted components ===
def compute_modes_from_peak(x, components):
    modes = []
    for component in components:
        peak_index = np.argmax(component)  # Find index of max value
        modes.append(x[peak_index])  # Store corresponding x-value as mode
    return modes


# === Monte Carlo simulation parameters ===
n_simulations = monte_carlo_simulations
sigma_noise = np.std(pdf - best_fit)

# Pre-allocate storage for Monte Carlo simulations
simulated_fits = np.zeros((n_simulations, len(x_log_uT)))
simulated_comps = np.zeros((n_simulations, n_active_components, len(x_log_uT)))
simulated_params = np.zeros((n_simulations, n_active_components * 4))

# Monte Carlo simulation loop
np.random.seed(77)  # Use any high value integer to ensure reproducibility (e.g. 77, 1234, 2025)

for i in range(n_simulations):
    noisy_y = pdf + np.random.normal(0, sigma_noise, size=pdf.shape)
    result_noisy = minimize(objective, optimized_params, args=(x_log_uT, noisy_y, n_active_components), method='L-BFGS-B')
    simulated_params[i, :] = result_noisy.x
    best_fit_noisy = model(x_log_uT, result_noisy.x, n_active_components)
    simulated_fits[i, :] = best_fit_noisy
    
    for j in range(n_active_components):
        amplitude = result_noisy.x[j * 4]
        center = result_noisy.x[j * 4 + 1]
        sigma = result_noisy.x[j * 4 + 2]
        skew = result_noisy.x[j * 4 + 3]
        simulated_comps[i, j, :] = skewed_gaussian(x_log_uT, amplitude, center, sigma, skew)

# Calculate the mean and standard deviation across simulations for fits and components
mean_fit = np.mean(simulated_fits, axis=0)
fit_std = np.std(simulated_fits, axis=0)

mean_comps = np.mean(simulated_comps, axis=0)
comps_std = np.std(simulated_comps, axis=0)

# Compute peak-based modes
modes = compute_modes_from_peak(x_log_uT, mean_comps)


#== Calculate PDF Residuals ==
pdf_residuals= pdf - best_fit
````

## Skewed Gaussian CDF Model Definition
$$
\text{CDF}(x) = A \left[ \frac{1}{2} \left(1 + \operatorname{erf}\left(\frac{x - \mu}{\sqrt{2} \sigma} \right) \right) - 2 \cdot T\left( \frac{x - \mu}{\sigma}, \alpha \right) \right]
$$
- $A$: Amplitude
- $\mu$: Center
- $\sigma$: Dispersion (standard deviation)
- $\alpha$: Skewness
- $\operatorname{erf}$: Error function
- $T(h, a)$: Owenâs T-function, defined as
$$
T(h, a) = \frac{1}{2\pi} \int_0^a \frac{e^{-\frac{1}{2} h^2(1 + t^2)}}{1 + t^2} , dt
$$
where
- $h = \frac{x - \mu}{\sigma}$: Standardized distance from the center
- $x$: Independent variable (log-scaled field, e.g., $\log_{10} B$)
- t: Dummy variable of integration



This model introduces asymmetry into the cumulative distribution through the Owenâs T-function, allowing for flexible skewed behavior around the center $\mu$.

The parameters $(A, \mu, \sigma, \alpha)$ used here are directly inherited from the optimized PDF fit. They are applied to the CDF to ensure consistent modeling across both domains. These same parameters are then used in Monte Carlo simulations to estimate confidence bounds on the CDF components, which may be overestimated due to the increased noise present in the gradient-based PDF data.

````python
#== Defining Skewed Gaussian CDF and Model Functions ==
def skewed_gaussian_cdf(x, amplitude, center, sigma, skew):
    skew= -skew
    erf_term_cdf = (1 + erf((x - center) / (np.sqrt(2) * sigma)))
    owen_term = 2 * owens_t((x - center) / sigma, skew)
    return amplitude * (0.5 * erf_term_cdf - owen_term)

# CDF model using centers
def cdf_model(x, params, n_components):
    y = np.zeros_like(x)
    for i in range(n_components):
        amp, cen, sig, skw = params[i*4:i*4+4]
        y += skewed_gaussian_cdf(x, amp, cen, sig, skw)
    return y


#== Monte Carlo simulations around CDF components ==
fitted_cdf = cdf_model(x_log_uT, optimized_params, n_active_components)
simulated_cdf_fits = np.zeros_like(simulated_fits)
simulated_cdf_comps = np.zeros_like(simulated_comps)

for i in range(n_simulations):
    for j in range(n_active_components):
        amp = simulated_params[i, j * 4]
        cen = simulated_params[i, j * 4 + 1]
        sig = simulated_params[i, j * 4 + 2]
        skw = simulated_params[i, j * 4 + 3]

        simulated_cdf_comps[i, j, :] = skewed_gaussian_cdf(x_log_uT, amp, cen, sig, skw)

    simulated_cdf_fits[i, :] = np.sum(simulated_cdf_comps[i], axis=0)

mean_cdf_fit = np.mean(simulated_cdf_fits, axis=0)
std_cdf_fit  = np.std(simulated_cdf_fits, axis=0)

mean_cdf_comps = np.mean(simulated_cdf_comps, axis=0)
std_cdf_comps  = np.std(simulated_cdf_comps, axis=0)


#== Calculate CDF Residuals ==
cdf_residuals = cdf - fitted_cdf
````

## Plotting Overview

### Data Types
- **CDF (Cumulative Distribution Function):** Original measurement data.
- **PDF (Probability Density Function):** Derived by computing the gradient of the CDF with respect to log B. Highlights individual components and inflection points more clearly.

### Fitting Strategy
- All **fitting, optimization**, and **Monte Carlo uncertainty estimation** are performed on the **PDF**, despite its higher noise levels.
- The optimized PDF parameters (amplitude, center, sigma, skewness) are then used to **fit the CDF**, with **amplitude readjustment only**, to ensure accurate scaling.
- As a result, **uncertainty envelopes** derived from the noisy PDF **may appear overestimated when applied to the CDF**.
- Fitted Mode (PDF) and Center (CDF) field values are reported in the plot legends.

### PDF Component Intersections
**PDF plots** feature vertical lines marking the **intersection points between adjacent components**, with the field values also listed in the legend. These values are useful for deriving rock-magnetic parameters (e.g., **s-ratios**, **HIRM**) and for conducting subsequent experiments such as the **Goethite Test** (**classic** and **HIRM-based**). Note that the algorithm may fail to return intersections when adjacent peaks are too close or poorly resolved.

## Optimized Parameter Printout and Save to CVS

````python
# Compose dynamic CSV output file name
output_file = f"{base_name}_optimized_components.csv"

# Write header and data to CSV
with open(output_file, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Component", "Amplitude", "Center (logB)", "Center (mT)",
                     "Mode (logB)", "Mode (mT)", "Sigma", "Skew"])

    for i in range(n_active_components):
        amplitude = optimized_params[i * 4]
        center_log_uT = optimized_params[i * 4 + 1]
        sigma = optimized_params[i * 4 + 2]
        skew = optimized_params[i * 4 + 3]
        mode_log_uT = modes[i]

        center_log_mT = center_log_uT - 3
        center_log_mT_rounded = round(center_log_mT, 2)
        center_mT = 10 ** center_log_mT_rounded

        mode_log_mT = mode_log_uT - 3
        mode_log_mT_rounded = round(mode_log_mT, 2)
        mode_mT = 10 ** mode_log_mT_rounded

        writer.writerow([
            f"Component {i + 1}",
            f"{amplitude:.8f}",
            f"{center_log_mT_rounded:.2f}",
            f"{center_mT:.2f}",
            f"{mode_log_mT_rounded:.2f}",
            f"{mode_mT:.2f}",
            f"{sigma:.2f}",
            f"{skew:.2f}"
        ])

        display(Markdown(
            f"**Component {i + 1}:** Amplitude= {amplitude:.8f}; "
            f"Center= {center_log_mT_rounded:.2f} (logB), {center_mT:.2f} (mT); "
            f"Mode= {mode_log_mT_rounded:.2f} (logB), {mode_mT:.2f} (mT); "
            f"Sigma= {sigma:.2f}; Skew= {skew:.2f}"
        ))
````
